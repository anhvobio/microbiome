#Create Manifest file

```python
# prompt: list name all files with fastq.gz in content/

import os
fastq_files = [f for f in os.listdir('/content/') if f.endswith('.fastq.gz')]
print(fastq_files)
```

```python
import pandas as pd

# Tạo danh sách các file forward
forward_files = fastq_files


# Tạo DataFrame
data = {'SampleID': [f"sample{i+1}" for i in range(len(forward_files))],
        'absolute-filepath': [f"/content/{file}" for file in forward_files]}
df = pd.DataFrame(data)

# Lưu vào file
df.to_csv("manifest.tsv", sep='\t', index=False)
```

#Denoise



##import

```python
%%shell
# Activate qiime2
source activate qiime2-amplicon-2024.5

# Import data to qiime2 aftifact
qiime tools import \
  --type "SampleData[SequencesWithQuality]" \
  --input-format  SingleEndFastqManifestPhred33V2 \
  --input-path /content/manifest.tsv \
  --output-path /content/demux_seqs.qza

# View quality of our samples
qiime demux summarize \
  --i-data /content/demux_seqs.qza \
  --o-visualization /content/demux_seqs.qzv
```

```python
%%shell
# Activate qiime2
source activate qiime2-amplicon-2024.5

# Import data to qiime2 aftifact
qiime metadata tabulate \
  --m-input-file /content/412_metadata.tsv \
  --o-visualization /content/metadata.qzv
```

##denoise dada2

```python
%%shell
# Activate qiime2
source activate qiime2-amplicon-2024.5

# Import data to qiime2 aftifact
qiime dada2 denoise-single \
  --i-demultiplexed-seqs /content/demux_seqs.qza \
  --p-trim-left 0 \
  --p-trunc-len 250 \
  --p-max-ee 2 \
  --o-representative-sequences rep-seqs.qza \
  --o-table table.qza \
  --o-denoising-stats denoising-stats.qza
```

```python
%%shell
# activate qiime2
source activate qiime2-amplicon-2024.5

# view dada2_stats
qiime metadata tabulate \
  --m-input-file /content/denoising-stats.qza \
  --o-visualization /content/denoising-stats.qzv

# view feature table
qiime feature-table summarize \
  --i-table /content/table.qza \
  --m-sample-metadata-file /content/412_metadata.tsv \
  --o-visualization /content/table.qzv
```

```python
%%shell
# Activate qiime2
source activate qiime2-amplicon-2024.5

#Summarize the Feature Table and Feature Data
qiime feature-table tabulate-seqs \
--i-data /content/rep-seqs.qza \
--o-visualization /content/rep-seqs.qzv
```

#Build phylogeny

```python
%%shell
# activate qiime2
source activate qiime2-amplicon-2024.5

wget \
  -O "sepp-refs-gg-13-8.qza" \
  "https://data.qiime2.org/2019.10/common/sepp-refs-gg-13-8.qza"

# build phylogeny tree
qiime fragment-insertion sepp \
  --i-representative-sequences /content/rep-seqs.qza \
  --i-reference-database /content/sepp-refs-gg-13-8.qza \
  --o-tree /content/tree.qza \
  --o-placements /content/tree_placements.qza \
  --p-threads 2
```

```python
%%shell
# activate qiime2
source activate qiime2-amplicon-2024.5
qiime diversity alpha-rarefaction \
  --i-table /content/table.qza \
  --m-metadata-file /content/412_metadata.tsv \
  --o-visualization /content/alpha_rarefaction_curves.qzv \
  --p-min-depth 10 \
  --p-max-depth 4250
```

#DAA

```python
%%shell
# activate qiime2
source activate qiime2-amplicon-2024.5

qiime feature-table filter-samples \
  --i-table /content/table.qza \
  --m-metadata-file /content/412_metadata.tsv \
  --p-where "[Location]='LGI'" \
  --o-filtered-table /content/LGI-table.qza
```

```python
%%shell
# activate qiime2
source activate qiime2-amplicon-2024.5

qiime taxa collapse \
  --i-table /content/table.qza \
  --i-taxonomy /content/taxonomy.qza \
  --p-level 6 \
  --o-collapsed-table /content/table-l6.qza

qiime composition ancombc \
  --i-table /content/table-l6.qza \
  --m-metadata-file /content/412_metadata.tsv \
  --p-formula 'Location' \
  --o-differentials /content/ancombc-location.qza

qiime composition da-barplot \
  --i-data /content/ancombc-location.qza \
  --p-significance-threshold 0.001 \
  --p-level-delimiter ';' \
  --o-visualization /content/da-barplot-colaction.qzv
```
